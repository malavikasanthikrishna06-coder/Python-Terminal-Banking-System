```python
import sys

class BankAccount:
    def __init__(self, account_holder, initial_balance=0.0):
        self.holder = account_holder
        self.balance = initial_balance
        self.transaction_history = [f"Account opened with balance: ${initial_balance:.2f}"]

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transaction_history.append(f"Deposited: ${amount:.2f}")
            print(f"\n✅ Success! Deposited ${amount:.2f}. New Balance: ${self.balance:.2f}")
        else:
            print("\n❌ Error: Deposit amount must be positive!")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"\n❌ Declined: Insufficient funds! Current balance is ${self.balance:.2f}")
        elif amount <= 0:
            print("\n❌ Error: Withdrawal amount must be positive!")
        else:
            self.balance -= amount
            self.transaction_history.append(f"Withdrew: ${amount:.2f}")
            print(f"\n✅ Success! Withdrew ${amount:.2f}. Remaining Balance: ${self.balance:.2f}")

    def display_statement(self):
        print(f"\n=== 📄 BANK STATEMENT FOR {self.holder.upper()} ===")
        for transaction in self.transaction_history:
            print(f" -> {transaction}")
        print(f"=====================================")
        print(f"FINAL NET BALANCE: ${self.balance:.2f}\n")


# --- RUNNING THE INTERACTIVE INTERFACE ---
print("🏦 Welcome to the Python Terminal Banking System")
name = input("Enter account holder name to start: ")
user_account = BankAccount(name, initial_balance=100.00) # Starts with $100

while True:
    print("\n--- AVAILABLE OPTIONS ---")
    print("1. Check Balance & Statement")
    print("2. Make a Deposit")
    print("3. Withdraw Cash")
    print("4. Exit System")

    choice = input("Select an option (1-4): ")

    if choice == '1':
        user_account.display_statement()
    elif choice == '2':
        amt = float(input("Enter deposit amount: $"))
        user_account.deposit(amt)
    elif choice == '3':
        amt = float(input("Enter withdrawal amount: $"))
        user_account.withdraw(amt)
    elif choice == '4':
        print(f"\n👋 Thank you for banking with us, {name}! Transaction session closed.")
        break
    else:
        print("\n❌ Invalid choice! Please select a number between 1 and 4.")
```
